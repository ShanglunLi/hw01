# Selecting and Fitting a Model

### Problem 1.

#### (a)

When the sample size $n$ is extremely large while the number of predictors $p$ is small, it would be better to use a flexible model, since we can fit in more parameters and we will have a small variance because of limited number of predictors.

#### (b)

When the number of predictors $p$ is extremely large while the sample size $n$ is small, it would be better to use an inflexible model. With a small sample size, we are more likely to encounter the overfitting problem when we use a flexible model.

#### (c)

When the relationship between the predictors and response is highly non-linear, it would be better to use a flexible model, since a flexible model can make the shape of hit less restrictive.

#### (d)

When the variance of the error terms is extremely high, it would be better to use an inflexible model. In this case, a flexible model may end up with overfitting problem.

### Problem 2.

#### (a)

```python
import numpy as np
import pylab 

x = np.linspace(0, 20, 1000)
y1 = 300 * np.cos(x / 3) + 500 + x**3 / 3
y2 = x**3 / 3
y3 = 0 * x + 250
y4 = 300 * np.cos(x / 3) + 350 
y5 = 225 * np.cos(x / 3) + 450 

pylab.plot(x, y1, '-b', label='Test Error')
pylab.plot(x, y2, '-r', label='Variance')
pylab.plot(x, y3, '-c', label='Irreducible Error')
pylab.plot(x, y4, '-m', label='Bias')
pylab.plot(x, y5, '-y', label='Train Error')
pylab.legend(loc='upper left')
pylab.ylim(0, 1000)
pylab.xlim(0, 10.0)
pylab.show()
```

#### (b)

Irreducible error is the same all the time.

The higher flexibility the less bias we have, since for higher flexibility model, there are fewer assumptions made about the shape of the fit. 

With higher flexibility, data will affect more on the parameter estimates. Thus, variance will increase with higher flexibility.

Test error will have a curved shape, since it is the combination of variance and bias.

Since an overfit model will cause lower MSE on the training data, training error will always decrease with more model flexibility.

### Problem 3.

#### (a)

```python
import numpy as np
np.random.seed(123)
```

#### (b)

```python
x1 = np.random.uniform(-1,1,200)
x2 = np.random.uniform(-1,1,200)
```

#### (c)

```python
e = np.random.normal(0, 0.5, 200)
y = x1 + x1 ** 2 + x2 + x2 ** 2 + e
```

#### (d)

```python
def odd(x):
    return x / (1+x)

v_odd = np.vectorize(odd)
p = v_odd(np.exp(y))
```

#### (e) - (h)

```python
import matplotlib.pyplot as plt
from numpy import arange
from numpy import meshgrid

delta = 0.01
xrange = np.arange(-1, 1, delta)
yrange = np.arange(-1, 1, delta)
X1, X2 = np.meshgrid(xrange,yrange)

fig, ax = plt.subplots()

group = np.array(['success' if i > 0.5 else 'failure' for i in p])
for g in np.unique(group):
    i = np.where(group == g)
    ax.scatter(x1[i], x2[i], label = g)
    
ax.legend()
plt.contour(X1, X2, np.exp(X1 + X1 ** 2 + X2 + X2 ** 2), [1])
plt.title('The Bayes Decision Boundary by Two Simulated Dataset')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.show()
```
